**Задача**

Вам необходимо оформить таблицу, которая станет основой рекомендательной системы.

Итоговая таблица должна состоять из трёх столбцов:

- Курс, к которому идёт рекомендация.
- Курс для рекомендации № 1 (самый популярный).
- Курс для рекомендации № 2 (второй по популярности).

```python
import pandas as pd
from collections import defaultdict
from collections import Counter
from itertools import combinations
```

Создадим словарь, где будем хранить в виде списка для одного пользователя пройденные им курсы.

```python
temp_dict = defaultdict(list)
for i in range(len(df_user)):
    temp_dict[df_user['user_id'][i]].append(df_user['resource_id'][i])

temp_dict 
```

Далее формируем максимально возможное количество пар среди курсов, которые проходили пользователи. Это делается для того, чтобы понять какая связка курсов наиболее встречаемая, а значит, с большей вероятностью user может совершить покупку. По этой связке мы будем предлагать дополнительный курс для пользователей.

```python

list_combinations = []
for i in temp_dict:
    for j in list(combinations(temp_dict[i],2)):
        list_combinations.append(j)

list_combinations
```

Считаем наиболее часто встречающися связки.

```python
amount_combinations = Counter(list_combinations).most_common()

amount_combinations
```

Подсчет количества связок.

```python
len(amount_combinations)
```

Отбрасываем связки, которые встречаются в единичных случаях.

```python
min_border = 10

amount_combinations_min_border = [amount_combinations[i] for i in range(len(amount_combinations)) if amount_combinations[i][1] >= min_border]

amount_combinations_min_border
```

```python
df_user = df_user.drop('user_id',axis=1)

df_user
```

```python
df_recommended_table = df_user.drop_duplicates()

df_recommended_table
```

```python
df_recommended_table.reset_index(drop=True, inplace=True)


df_recommended_table
```

Теперь для каждого отдельного курса ищем 2 самых часто встречающиеся связки с участием данного курса. Их будем рекомендовать пользователю.

Алгоритм:
Для каждого курса (resource_id) из датафрейма df_recommended_table будем проверять, есть ли данный курс в связке из amount_combinations_min_border. Связки пар расположены в порядке убывания по количеству встречаемости в истории покупок клиентов. Если такой курс есть в какой-то связке, то мы проверяем первая или вторая рекомендация для этого курса и вытаскиваем второе значение из данной связки пар и записываем его в рекомендацию.

Могут попадаться такие курсы, для которых вторая пара встречалась меньше 10 раз. У таких курсов в рекомендациях обозначено None. В моем случае в качестве рекомендации к таким курсам, будет предлагаться рандомно любые 1 или 2 курса. Кто знает, может новая пара покажет себя очень хорошо с точки зрения прибыли.


```python
df_recommended_table['Recommendation_1'] = None
df_recommended_table['Recommendation_2'] = None
pd.options.mode.chained_assignment = None

for i in range(len(df_recommended_table)):
    count = 0
    val = df_recommended_table['resource_id'][i]
    for combination in amount_combinations_min_border:
        if val in combination[0]:
            if not count:
                for elem in combination[0]:
                    if elem != val:
                        df_recommended_table['Recommendation_1'][i] = elem
                        count+=1
            else:
                for elem in combination[0]:
                    if elem != val:
                        df_recommended_table['Recommendation_2'][i] = elem
                break
```

```python
import random
```

```python
list_course = list(df_recommended_table['resource_id'])
```

Алгоритм:
Проверяем, есть ли значение None в столбцах Recommendation_1 и Recommendation_2 в текущей строчке датафрейма df_recommended_table. Если есть в одном из этих столбцов значение None, то берем рандомное значения из списка всех курсов. При этом проверяем совпало ли рандомное значение с текущим значением resource_id. Если совпало, то заново рандомно берем значение из списка курсов до тех пор пока получившееся значение будет отличаться от текущего значения resource_id.

```python
for i in range(len(df_recommended_table)):
    val = df_recommended_table['resource_id'][i]
    if df_recommended_table['Recommendation_1'][i] is None and df_recommended_table['Recommendation_2'][i] is None:
        randomize = random.sample(list_course,2)
        while (val in randomize):
            randomize = random.sample(list_course,2)
        df_recommended_table['Recommendation_1'][i] = randomize[0]
        df_recommended_table['Recommendation_2'][i] = randomize[1]
        continue
        
    if df_recommended_table['Recommendation_1'][i] is None:
        randomize = random.sample(list_course,1)
        while (randomize[0]==val):
            randomize = random.sample(list_course,1)
        df_recommended_table['Recommendation_1'][i] = randomize[0]
        continue
    if df_recommended_table['Recommendation_2'][i] is None:
        randomize = random.sample(list_course,1)
        while (randomize[0]==val):
            randomize = random.sample(list_course,1)
        df_recommended_table['Recommendation_2'][i] = randomize[0]
        
df_recommended_table
```
